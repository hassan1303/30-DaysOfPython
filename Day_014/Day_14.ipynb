### Question 1

```python
# map 
'''
    The map() function executes a specified function for 
    each item in an iterable. The item is sent to the function as a 
    parameter.
'''   
# filter
''''The filter() function returns an iterator were the 
items are filtered through a function to test if the item 
is accepted or not.
'''

# reduce
'''The reduce() function is used when you have list of values, 
but you want to use the values in that list to create something 
completely new. 
'''
```

### Question 2

```python
# higher order function
'''A higher order function is a function that takes a function as an argument
'''
# decorator
'''A decorator in Python is an example of a higher-order function, but there are 
decorators that are not and there are higher-order functions that are not decorators,
for example those that take two required arguments that are functions.
'''
# Closure
'''A Closure is a function object that remembers values in enclosing scopes
even if they are not present in memory.
'''
```

### Question 3

```python
def square(num):
    return num ** 3


def long_name(name):
    if len(name) >= 7:
        return True
    return False


print(list(filter(long_name, names)))
print(list(map(square, numbers)))
```

```python
keys = []
keys = [i[0] for i in all_countries if i[0] not in keys]

def country_name_counter(same_name):
    return sum([True for j in all_countries if j[0].startswith(same_name)])


values = [country_name_counter(i) for i in keys]
country_dict = dict(zip(keys,values))
print(country_dict)
```
